```python
import pandas as pd
```

```python
df=pd.read_csv("../data/processed/final_merged_file.csv")
df.head(5)
```

```python
df.columns
```

```python
df['match_date'] = pd.to_datetime(df['match_date'])
```

```python
df['birthday'] = pd.to_datetime(df['birthday'])
```

```python
#How many matches does each player play per season
df.groupby(["id","player_name","season"]).agg(match_play_season=("season","count")).reset_index()
```

```python
#- What is the average number of matches per month/week per player?
#matches per month
df["Match_month"]=df["match_date"].dt.month
```

```python
match_per_month=df.groupby(["Match_month","player_name"]).agg(monthly_count=("id","count")).reset_index()
```

```python
#avg number of match per month player palays
match_per_month.groupby("player_name").agg(avg_player_match_month=("monthly_count","mean")).reset_index().sort_values(by="avg_player_match_month",ascending=False)
```

```python
#per week
df["match_week"]=df["match_date"].dt.isocalendar().week
```

```python
player_match_week=df.groupby(["match_week","player_name"]).agg(weekly_count=("id","count")).reset_index()
player_match_week.groupby("player_name").agg({"weekly_count":"mean"})
```

```python
#Which players have the highest total matches across all seasons?
df.groupby(["player_name","season"]).agg(total_match=("id","count")).reset_index().sort_values(by="total_match",ascending=False)
```

```python
#- How does workload vary across leagues (league_id) and seasons?
df.groupby(["league_id", "season"]).agg(matches_played=("id", "count")).reset_index().sort_values(by="matches_played", ascending=False)
```

```python
#Do players with higher match counts per season show lower stamina ratings?
df.groupby(["player_name","season"]).agg({"id":"count","stamina":"mean"}).reset_index().sort_values(by="id",ascending=False).head(10)
```

```python
#- Players with very high workloads (80+ matches) often show stamina ratings in the 60–70 range, but some drop much lower (Ricardo, Rafael).
#- There’s variation: some players maintain high stamina (Fernando ~78.8, Maicon ~76.4) even with heavy workloads.
#It shows that while many high‑workload players hover around stamina ~65–75, some drop sharply, hinting at fatigue effects.
```

```python
#- Is there a noticeable drop in stamina for players who consistently play more matches per month?
df.groupby("player_name").agg({"Match_month":"count","stamina":"mean"}).reset_index().sort_values(by="Match_month",ascending=False)
```

```python
#- Does sprint speed decline for players with heavier workloads?
workload_attributes=df.groupby(["player_name","season"]).agg({"id":"count","sprint_speed":"mean"}).reset_index().sort_values(by="id",ascending=False).head(10)
workload_attributes
```

```python
#- Some high‑workload players (Ricardo, Felipe, Rafael) show much lower sprint speed.
# Others (Rafinha, Maicon, Marcelo) maintain strong sprint speed despite heavy workloads.
```

```python
correlations = workload_attributes[["id","sprint_speed"]].corr()
print(correlations)
```

```python
#correlation result shows a weak negative relationship between workload and sprint speed. It hints at fatigue effects but isn’t strong enough to prove
# average sprint speed rating decreases as match workload increases
```

```python
df["age"]=(df["match_date"]-df["birthday"]).dt.days // 365
```

```python
df["age"].unique()
```

```python
#- Do younger players maintain stamina better than older players under similar workloads?

workload_stamina=df.groupby(["player_name","season","age"]).agg({"id":"count","stamina":"mean"}).reset_index().sort_values(by="stamina",ascending=False)
```

```python
workload_stamina["age-segment"]=workload_stamina["age"].apply(lambda X: 'younger'if X <25 else 'older')
```

```python
workload_stamina.groupby(["id","age-segment"]).agg({"stamina":"mean"}).reset_index().sort_values(by="stamina",ascending=False).head(10)
```

```python
#- There isn’t a dramatic difference between younger and older players in this slice of data.
# Younger players slightly edge out older ones at the top end (83.9 vs 83.1), but overall the ranges overlap heavily.
#shows that both younger and older players can maintain high stamina under workload, with only a small advantage for younger players
```

```python

# 4. How does age (derived from birthday) affect overall_rating and potential?
df.groupby("age").agg({"overall_rating":"mean","potential":"mean"}).reset_index().sort_values(by=["overall_rating","potential"],ascending=False)
#Younger players = low rating, high potential
#Prime-age players (28–33) = highest ratings
#Older players = high rating but low potential
```

```python
# 6. Do players show performance decline (overall_rating drop) as the season progresses (stage increases)?
df.groupby("stage").agg({"overall_rating":"mean"}).reset_index().sort_values(by="overall_rating",ascending=False)
#Players do NOT show a clear performance decline as the season progresses.
# Average overall_rating stays almost constant across all stages.
# Any small differences are random noise, not fatigue effects
#Fatigue does not appear in overall_rating at the season level.
```
