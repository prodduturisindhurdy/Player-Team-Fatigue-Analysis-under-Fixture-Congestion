cleaning player attributes file

```python
players_att=pd.read_csv("../data/raw/Player_Attributes.csv")
```

keeping only required columns

```python
players_att=players_att[["player_api_id","player_fifa_api_id","date","overall_rating","potential","stamina","strength","jumping",
"balance","acceleration","sprint_speed","agility","reactions","interceptions","aggression","positioning"]]
```

replacing null values with the median values for null numeric columns

```python
columns=["overall_rating","stamina","potential",'strength', 'jumping', 'balance',
       'acceleration', 'sprint_speed', 'agility', 'reactions', 'interceptions',
       'aggression', 'positioning']
for col in columns:
    players_att[col].fillna(players_att[col].median(),inplace=True)
```

```python
players_att.isna().sum()
```

```python
players_att.dtypes
```

```python
players_att["date"]=pd.to_datetime(players_att["date"])
```

```python
players_att["date"]
```

merging three tables

```python
merged_df = pd.merge(
    left=players,
    right=players_att,
    how="inner",
    left_on="player_api_id",
    right_on="player_api_id"
)
merged_df
final_merged=pd.merge(merged_df,match,on="id")
```

```python
final_merged=final_merged.rename(columns={"player_fifa_api_id_x":"player_fifa_api_id","date_x":"rating_date","date_y":"match_date"})
```

description


```python
# id: Unique identifier for the player record (from player table)
# player_api_id: Internal ID linking player across tables (player, player_attributes, match)
# player_name: Players full name
# player_fifa_api_id: FIFA database ID for the player (external reference)
# birthday: Players date of birth (datetime64)  used to calculate age at match
# height: Players height in centimeters
# weight: Players weight in kilograms
# player_fifa_api_id_y: Duplicate FIFA ID from player_attributes table (can be dropped)
# rating_date: Date when player attributes (ratings) were recorded in player_attributes
# overall_rating: General performance rating (FIFA-style)
# potential: Expected maximum rating potential of the player
# stamina: Endurance level; key fatigue indicator
# strength: Physical power attribute
# jumping: Ability to jump; aerial performance
# balance: Stability and control attribute
# acceleration: Speed of reaching top pace
# sprint_speed: Maximum sprinting speed
# agility: Flexibility and quickness in movement
# reactions: Response time to events in the game
# interceptions: Defensive ability to intercept passes
# aggression: Players assertiveness and intensity
# positioning: Ability to position correctly during play
# match_date: Date of the match (from match table)
# season: Season identifier (e.g., "2012/2013")
# stage: Stage of the competition (league round or knockout stage)
# league_id: ID of the league in which the match was played
# country_id: ID of the country where the league is based
# home_team_api_id: ID of the home team
# away_team_api_id: ID of the away team
# home_team_goal: Number of goals scored by the home team
# away_team_goal: Number of goals scored by the away team
```
